The worst month during this period os May, followed by June, March, April, January and February. Let's take a look the which time during the day is the worst.

````python
data['HORA'].value_counts()
````

Need to strip time only by hour.

````python
import time

hour_buffer = []
for x in data['HORA']:
    if len(x) >= 7:
        time_obj = time.strptime(x, '%H:%M:%S')
        hour_buffer.append(time_obj.tm_hour)
    else:
        time_obj = time.strptime(x, '%H:%M')
        hour_buffer.append(time_obj.tm_hour)
            
data['HOUR_ONLY'] = hour_buffer
````

````python
data['HOUR_ONLY'].hist(bins=50, figsize=(8,5))
plt.show()
````

## How the gunfires look like in the map

Let's see how the reports look like on the map. First, we have to performe some cleaning on the dataset. The module ```Cleaning``` has the
transformes ```RemoveOutliers_lat_long``` and ```String2Float_lat_long```, that will remove outliers far from Rio de Janeiro city and convert the values of Latitude and Longitude of string into float.

````python
from src.data import Cleaning

clean_01 = Cleaning.RemoveOutliers_lat_long()
clean_02 = Cleaning.String2Float_lat_long()

data = clean_01.fit_transform(data)
data = clean_02.fit_transform(data)
````

That's how it looks on the map of the State.

````python
import matplotlib.image as mpimg
rio_img = mpimg.imread('../figs/RJ_state.png')
ax = data.plot(kind='scatter', x='LONGITUDE', y= 'LATITUDE', figsize=(15,10), alpha = 0.5, s=50, c='r')
plt.imshow(rio_img, extent=[-44.93, -40.81, -23.43, -20.72], alpha=1, cmap=plt.get_cmap("jet"))
plt.show()
````

There is such small data outside from the center of the city. For now, let's focus on the region around Rocinha. To make it easier there is a transformer named ```GetData_lat_long```, which gets only the points of data on the give tuple.

````python
    clean_03 = Cleaning.GetData_lat_long(coordinates=(-43.5, -43.2, -23.05, -22.8))
    data_selected = clean_03.fit_transform(data)
````

````python
rio_city_img = mpimg.imread('../figs/RJ_city.png')
plt.figure(figsize=(18,10))
plt.imshow(rio_city_img, extent=(-43.906927, -42.987857, -23.081902, -22.736000), alpha=1, cmap=plt.get_cmap("jet"))
plt.scatter(data_selected['LONGITUDE'], data_selected['LATITUDE'], c='r')
plt.show()
````

I'm curious to see how the "top10" looks like. Let's take a look.

The ```groupby``` method of pandas might come in handy. Let's take a mean value of coordinates for each neighborhood.

````python
bairros_coord = data.groupby(by='BAIRRO')['LATITUDE', 'LONGITUDE'].agg(['mean'])
bairros_coord.head()
````

Now, get the report on each neighborhood

````python
bairros_count = data.groupby(by='BAIRRO')['BAIRRO'].agg(['count'])
bairros_count.head()
````

````python
subset_data_bairros = pd.concat([bairros_coord, bairros_count], axis=1)
subset_data_bairros.columns = ['LATITUDE', 'LONGITUDE', 'GUNFIRE']
subset_data_bairros.head()
````

### The Top10 is:

````python
top10 = subset_data_bairros.sort_values(by='GUNFIRE', ascending=False).iloc[0:10]
top10
````

After some manipulation let's see how it looks.

````python
plt.figure(figsize=(25,6))
plt.imshow(rio_city_img, extent=(-43.906927, -42.987857, -23.081902, -22.736000), alpha=1, cmap=plt.get_cmap("jet"))
sc = plt.scatter(top10['LONGITUDE'], top10['LATITUDE'], c=top10['GUNFIRE'], 
                 s=top10['GUNFIRE']*40, cmap=plt.get_cmap("Wistia"), alpha=0.5)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Reports density of top 10 neighborhoods', fontsize=20)
cbar = plt.colorbar(sc)
cbar.set_label('Number of reports')
plt.show()
````
